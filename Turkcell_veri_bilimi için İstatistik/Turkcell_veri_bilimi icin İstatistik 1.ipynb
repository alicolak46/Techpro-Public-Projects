# Rassal Değişkenler ve Olasılık Dağılımları

Değerlerini bir deneyin sonuçlarından alan değişkene rassal değişken denir.

# Bernoulli Dağılımı

Başarılı-başarısız, olumlu-olumsuz şeklinde iki sonuçlu olaylar ile ilgilenildiğinde kullanılan kesikli olasılık dağılımıdır.
Bernoulli dağılımı binom dağılımının özel bir halidir.




${\displaystyle f(k;p)= p^{k}(1-p)^{1-k}\quad {\text{for }}k\in \{0,1\}}$

${\displaystyle \operatorname {E} \left(X\right)=p}$

$\operatorname {Var} [X]=pq=p(1-p)$ 

```python
from scipy.stats import bernoulli
```

```python
p = 0.6
```

```python
rv = bernoulli(p)
rv.pmf(k=0)
```

```python
rv.pmf(k=1)
```

# Büyük sayılar Yasası 

Bir rassal değişkenin uzun vadeli kararlılığını tanımlayan olasılık teoremidir.

```python
import numpy as np
rng = np.random.RandomState(123)
for i in np.arange(1,21):
    deney_sayisi = 2**i
    yazi_turalar = rng.randint(0,2,size=deney_sayisi)
    yazi_olasiliklari = np.mean(yazi_turalar)
    print("Atış Sayısı:",deney_sayisi,"---" "Yazı Olasılığı: %.2f" % (yazi_olasiliklari*100))
```

# Binom Dağılımı

Bağımsız n deneme sonucu k başarılı olma olasılığı ile ilgilenildiğinde kullanılan dağılımdır.







## İş Uygulaması: Reklam Harcaması Optimizasyonu

**Problem:**
Çeşitli mecralara reklam veriliyor, reklamların tıklanma ve geri dönüşüm oranları optimize edilmeye çalışılıyor. Buna yönelik belirli bir mecrada çeşitli senaryolara göre reklama tıklama olasılıkları hesaplanmak isteniyor.

<font color=red>**Detaylar:**</font>
- Bir mecrada reklam verilecek
- Dağılım ve reklama tıklama olasılığı biliniyor(0.01)
-**Soru** : Reklamı 100 kişi gördüğünde 1,5,10 tıklanması olasılığı nedir?

```python
from scipy.stats import binom
```

```python
p = 0.01 
n= 100
rv = binom(n,p)
print(rv.pmf(1))
print(rv.pmf(5))
print(rv.pmf(10))
```

## İş Uygulaması: İlan Girişi Hata Olasılıkları

**Problem: **

Hatalı ilan girişi olasılıkları hesaplanmak isteniyor.

**Detaylar:**

- Bir yıl süresince ölçümler yapılıyor

- Dağılım biliniyor(Poisson) ve Lambda 0.1(ortalama hata sayısı)

- Hiç hata olmaması, 3 hata olması ve 5 hata olması olasılıkları nedir?

```python
from scipy.stats import poisson

lambda_ = 0.1
```

```python
rv = poisson(mu=lambda_)
print(rv.pmf(k=0))
```

```python
print(rv.pmf(k=3))
print(rv.pmf(k=5))
```

## İş Uygulaması: Satış Olasılıklarının Hesaplanması

**Problem:** Bir yatırım/toplantı öncesinde gelecek ay ile ilgili satışların belirli değerlerde gerçekleşmesi olasılıkları belirlenmek isteniyor.

**Detaylar:** 
- Dağılımın normal olduğu biliniyor
- Aylık ortalama satış sayısı 80K, standart sapması 5K
- 90K'dan fazla satış yapma olasılığı nedir?

90K'dan fazla olma olasılığı nedir?



70K'dan fazla olma olasılığı nedir?


73K'dan az olma olasılığı nedir?


Satışların 85K ile 90K arasında olma olasılığı nedir?


```python
from scipy.stats import norm
```

```python
#90'dan fazla olması
1- norm.cdf(90,80,5)
```

```python
#70'dan fazla olması
1- norm.cdf(70,80,5)
```

```python
#73'den az olması
norm.cdf(73,80,5)
```

```python
#85 ile 90 olması
norm.cdf(90,80,5)- norm.cdf(85,80,5)
```

- Uygun olasılık fonksiyonları ile olasılık hesapları yapıyoruz.
