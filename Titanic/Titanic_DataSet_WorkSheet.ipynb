```python
df.groupby(['Pclass', 'Sex'])
# Pclass ve Cinsiyeti Gruplandırdık Ama Neye Göre, Nasıl Bir Gruplandırma Yapacağımızı Söylemediğimiz İçin Sonuç Döndürmedi.
```

```python
df.groupby(['Pclass', 'Sex'])['Survived'].mean()

# Bilet Sınıflarına Göre Cinsiyetleri Gruplandırdı ve Hayatta Kalanların Ortalamasını Aldı.

# Örn: 1. Sınıfta Yer Alan Kadınların Hayatta Kalma Oranı %97, Erkeklerin Hayatta Kalma Oranı %37
```

```python
df.groupby(['Pclass', 'Sex'])['Age'].median()

# Medyan en çok tekrar eden değer demektir.

# Daha Yaşlılar Daha İyi Sınıfı Tercih Ediyor Sonucunu Çıkarabiliriz.
```

```python
df.groupby(['Pclass', 'Sex'])['Age'].mean()

# Yaş Ortalaması Arttıkça Daha İyi Sınıfın Tercih Edilmesi de Artıyor.
```
