```python
kill=pd.read_csv('PoliceKillingsUS.csv',encoding="windows-1252")
```

## 1a: Box plot of age in both genders separeted by manner of death.

```python
kill.manner_of_death.unique()
```

#### INSTRUCTION: Demonstrate the Qurtiles of age in both genders separeted by manner of death.**

```python
kill.dropna(inplace=True)
```

```python
plt.figure(figsize=(4.3, 3))

plt.tick_params(axis='x', labelsize=7)
plt.xlabel("gender", fontsize=7)

plt.tick_params(axis='y', labelsize=7)
plt.ylabel("age", fontsize=7)

ax = sns.boxplot(x="gender", y="age", data=kill , hue=kill.manner_of_death,
            flierprops=dict( markersize=4), boxprops=dict(linewidth=1.2))

for line in ax.lines:
    line.set_alpha(0.6)

legend = plt.legend(title= "manner_of_death",fontsize=7)
legend.get_title().set_fontsize('7');
```

```python
# with seaborn
```

## 1b: Swarm Plot 

#### INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python
plt.figure(figsize=(4.3, 3))

plt.tick_params(axis='x', labelsize=7)
plt.xlabel("gender", fontsize=7)

plt.tick_params(axis='y', labelsize=7)
plt.ylabel("age", fontsize=7)

sns.swarmplot(x="gender", y="age", data=kill , hue=kill.manner_of_death,size=3.8)

legend = plt.legend(title= "manner_of_death",fontsize=7)
legend.get_title().set_fontsize('7');
```

```python
# with seaborn
```

## 2: Plot 12 most dangerous cities

```python
kill.city.value_counts().head(12)
```

```python
total_numbers = kill.city.value_counts().head(12).to_numpy()
total_numbers
```

```python
city_names = kill.city.value_counts().head(12).index.to_numpy()
city_names
```

```python
plt.figure(figsize=(8.7, 4.2))

plt.tick_params(axis='x', labelsize=7)

plt.tick_params(axis='y', labelsize=7)

plt.xticks(rotation=45, color='b')

plt.title('Most Dangerous Cities', color='r', fontsize=10)

sns.barplot(city_names, total_numbers);
```

```python
# with seaborn
```

```python
kill.city.value_counts().head(12)
```

```python
city_names = kill.city.value_counts().head(12).index.to_numpy()
city_names
```

```python
plt.figure(figsize=(5, 3))

plt.tick_params(axis='x', labelsize=7)
plt.xlabel("city", fontsize=7)

plt.tick_params(axis='y', labelsize=7)
plt.ylabel("count", fontsize=7)

plt.xticks(rotation=45)

plt.title('Most Dangerous Cities', color='b', fontsize=10)

sns.countplot(data=kill,x= "city", order=kill.city.value_counts().iloc[:12].index);
```

```python
# with seaborn countplot
```
