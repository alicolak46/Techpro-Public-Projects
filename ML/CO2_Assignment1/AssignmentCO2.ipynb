### DETERMINE THE FEATURES AND TARGET

```python
X = df[["Engine Size(L)","Fuel Consumption City (L/100 km)","Fuel Consumption Hwy (L/100 km)","Fuel Consumption Comb (L/100 km)"]]
y = df["CO2 Emissions(g/km)"]
```

### TRAIN-TEST-SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
df.shape # We can use 0.1 for test,i think there is enough data for test 
```

```python
X_train, X_test , y_train , y_test = train_test_split(X,y,test_size = 0.1, random_state=101)
```

### SCALING THE DATA

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train) #SCALE THE FEATURES FOR TRAIN
```

```python
X_test_scaled = scaler.transform(X_test) #SCALE THE FEATURES FOR TEST BUT DO NOT FIT (DATALEAK)
```

### LINEAR REGRESSION MODEL

```python
MLRmodel = LinearRegression()
```

```python
MLRmodel.fit(X_train_scaled,y_train)
```

```python
y_pred = MLRmodel.predict(X_test_scaled)
y_train_pred = MLRmodel.predict(X_train_scaled)
```

```python
def train_val(y_train, y_train_pred, y_test, y_pred, i): 
    
    scores = {i+"_train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),                          
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
    i+"_test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
mls = train_val(y_train, y_train_pred, y_test, y_pred, "Multiplelinear") 
mls
```

```python
#let's define a simple function to see gridsearch results
def display_gridsearch_results(grid_model):

    print("Best Params ", grid_model.best_params_)
    print("Best Score: ", grid_model.best_score_)
    print("Cv Count: ", grid_model.cv)
```

```python
display_gridsearch_results(grid_model)
```

```python
#MODEL OLUÅžTURMA - FÄ°T ETME - DÄ°SPLAY ETME FONKSÄ°YONLARINI BERABER Ã‡ALIÅžTIRACAK BÄ°R FONKSÄ°YON YAZALIM

def perform_gridsearch(X_train_scaled, y_train, estimator, param_grid, cv=10, verbose=2):

    # GridSearchCV modelini oluÅŸtur
    grid_model = GridSearchCV(estimator=estimator,
                              param_grid=param_grid,
                              scoring="neg_root_mean_squared_error",
                              cv=cv,
                              verbose=verbose)

    grid_model.fit(X_train_scaled, y_train)

    # SonuÃ§larÄ± gÃ¶ster
    print("Best Params: ", grid_model.best_params_)
    print("Best Score: ", grid_model.best_score_)
    print("CV Count: ", grid_model.cv)
```

```python
perform_gridsearch(X_train_scaled,y_train,elastic_model,param_grid,cv=10,verbose=0)
```
