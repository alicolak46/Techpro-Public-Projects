```python
# Her bir Ã¼lkenin toplam izlenme sayÄ±sÄ±nÄ± bulalÄ±m
ulke_toplam_izlenme = df['country'].value_counts().reset_index()
ulke_toplam_izlenme.columns = ['country', 'Toplam_Izlenme']

# Toplam izlenme sayÄ±sÄ±na gÃ¶re sÄ±ralayalÄ±m
ulke_toplam_izlenme = ulke_toplam_izlenme.sort_values(by='Toplam_Izlenme', ascending=False)

# En Ã§ok izlenen 10 Ã¼lkeyi seÃ§elim
en_cok_izlenen_10_ulke = ulke_toplam_izlenme.head(10)

# Veriyi gÃ¶rselleÅŸtirelim
plt.figure(figsize=(10, 6))
plt.barh(en_cok_izlenen_10_ulke['country'], en_cok_izlenen_10_ulke['Toplam_Izlenme'], color='red' )
plt.title('En Ã‡ok Ä°zlenen 10 Ãœlke')
plt.xlabel('Ãœlke')
plt.ylabel('Toplam Ä°zlenme SayÄ±sÄ±')
plt.tight_layout()
plt.show()
```

```python
df['country'] = df['country'].astype(str)

plt.subplots(figsize=(25,15))
wordcloud=WordCloud(
background_color="white",
    width=1920,
    height=1080
).generate(" ".join(df.country))
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig("country.png")
plt.show()
```

```python
genre_new = df.listed_in.str.split(', ', expand = True).stack()
genre_new
```

```python
genre_new.value_counts()
```

```python
genre = genre_new.value_counts().reset_index()
genre.columns = ["kind", "cnt"]
genre = genre. sort_values(by="cnt", ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(genre.kind, genre.cnt , palette="autumn_r")
plt.title("En Ã§ok izlenen 10 tÃ¼r")
plt.xlabel("TÃ¼r")
plt.ylabel("Toplam izlenme sayÄ±sÄ±")
plt.xticks(rotation=90)
plt.tight_layout;
```

```python
fig, ax = plt.subplots()
ax = sns.countplot(data = df, x = "type",palette="autumn_r")
ax.set_title("Distribution of Type", color= "red")

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.3, p.get_height()+0.6));
```

```python
df.type.value_counts()
```

```python
x =df.type.value_counts().values
mylabels= df.type.value_counts().index

plt.figure(figsize=(10,8))
plt.title("Distribution of Type", color= "red")
plt.pie(x, labels=mylabels, labeldistance= 0.5, autopct= "%.2f",
        startangle=90, shadow=True, explode = [0.1,0])
plt.legend();
```
