```python
valid_OS = ["FreeDOS","Windows 11", "Windows 10","MacOS","Linux" ]

# Eğer Screen_Size Freedos,FreeDos,FDOS,DOS,Dos,Fdos ise yeni değerini FreeDOS yap
df.loc[df['OS'].isin(["Freedos","FreeDos","FDOS","DOS","Dos","Fdos"]), 'OS'] = "FreeDOS"

# Eğer OS W11,Windows,Win11,Win 11,w11,WIN,11 Home ise yeni değerini Windows 11 yap
df.loc[df['OS'].isin(["W11","Windows","Win11","Win 11","w11","WIN","11 Home" ]), 'OS']= "Windows 11"

# Eğer OS W10 ise yeni değerini Windows 10 yap
df.loc[df['OS'].isin(["W10","Win10","Win 10","w10","win10" ]), 'OS'] = "Windows 10"

# Eğer OS Mac ise yeni değerini MacOS yap
df.loc[df['OS'].isin(["Mac"] ), 'OS'] = "MacOS"

# Eğer Ubuntu,Linux,Lin,LİNUX ise yeni değerini Linux yap
df.loc[df['OS'].isin(["Ubuntu","Linux","Lin","LİNUX"]), 'OS'] = "Linux"

# Eğer Screen_Size valid_OS listesinde değilse, o zaman Windows 11 yap
df.loc[~df['OS'].isin(valid_OS), 'OS'] = "Windows 11"
```

```python
# İki boşluğu olanların ilk boşluğunu kaldır
df['Processor'] = df['Processor'].apply(lambda x: x.replace(" ", "", 1) if x.count(" ") >= 2 else x)

df['ProcessorType'] = df['Processor'].str.split('[- ]', 1, expand=True)[0]
df[ 'ProcessorModel'] = df['Processor'].str.extract(r'(\d{4}\d?\w+|\w\d{3}\d?|d{4})')[0]

# İlk sütunu büyük harfe dönüştürme (örneğin, "I5" yerine "i5" gibi)
df['ProcessorType'] = df['ProcessorType'].str.upper()
df.sample(50)
```

```python
df.ProcessorModel.isnull().sum()
```

```python
df[df.ProcessorModel.isnull()].sample(60)
```

```python
# Eğer ProcessorModel R0030NTi7 ise yeni değerini 13 yap
df.loc[df['ProcessorModel'].isin(["R0030"] ), 'ProcessorModel'] = "13"

# ProcessorModel sütunu NaN olan satırları bulun
nan_rows = df[df['ProcessorModel'].isna()]

# product_link sütununu kontrol ederek ProcessorModel sütununu güncelleyin
for index, row in nan_rows.iterrows():
    if 'mac' in row['product_link'].lower():
        df.at[index, 'ProcessorModel'] = "12"   #bunun yerine None yazılabilirdi ama int yapmak için yazdık
    else:
        df.at[index, 'ProcessorModel'] = "13"
        
# burada macbook olanlara karar verilmeli biz 12 atadık
```

```python

# Kalan null değerleri 11 ile doldur.
#df['ProcessorModel'].fillna("11", inplace= True)
```

```python
df.ProcessorModel.isnull().sum()
```

```python
# Generation sütununu oluşturma
df['ProcessorGeneration'] = df['ProcessorModel'].apply(lambda x: 
    int(x[1]) if x and x[0].isdigit() == False and len(x) > 1 and x[1].isdigit() 
    else int(x[:2]) if x and x[:2].isdigit() and int(x[:2]) < 15 
    else int(x[0]) if x and x[:2].isdigit() and int(x[:2]) > 15 
    else None
)
```
